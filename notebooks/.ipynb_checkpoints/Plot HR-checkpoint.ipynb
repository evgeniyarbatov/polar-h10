```python
import os
import gpxpy

import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.dates as mdates
```

```python
def read_gpx_to_df(file_path):
    with open(file_path, 'r') as gpx_file:
        gpx = gpxpy.parse(gpx_file)
        data = []
        for track in gpx.tracks:
            for segment in track.segments:
                for point in segment.points:
                    hr = next((ext.text for ext in point.extensions if ext.tag.endswith('hr')), None)
                    data.append({
                        'time': point.time,
                        'hr': float(hr) if hr else None,
                        'file_path': file_path,
                    })
            
        df = pd.DataFrame(data)
        
        df = df.dropna(subset=['hr'])
        df['time'] = pd.to_datetime(df['time']).dt.floor('s')
        df['time'] = df['time'].dt.tz_localize(None)

        df = df.set_index('time')
        df = df.resample('min').apply(
            lambda group: group.iloc[0] if group.name == 'file_path' else group.mean()
        )
        df = df.reset_index()
        
        start_time = df['time'].min()
        df['seconds_since_start'] = (df['time'] - start_time).dt.total_seconds().astype(int)

        df['timestamp'] = pd.to_timedelta(df['seconds_since_start'], unit='s').apply(lambda x: '{:02}:{:02}:{:02}'.format(int(x.total_seconds() // 3600), int((x.total_seconds() % 3600) // 60), int(x.total_seconds() % 60)))
        df['timestamp'] = pd.to_datetime(df['timestamp'], format='%H:%M:%S').dt.time

        df = df.drop(['time', 'seconds_since_start'], axis=1)
        return df

with_files = [os.path.join('../data/with', f) for f in os.listdir('../data/with') if f.endswith('.gpx')]
without_files = [os.path.join('../data/without', f) for f in os.listdir('../data/without') if f.endswith('.gpx')]

dfs_with = [read_gpx_to_df(f) for f in with_files]
dfs_without = [read_gpx_to_df(f) for f in without_files]

df_with = pd.concat(dfs_with, ignore_index=True)
df_without = pd.concat(dfs_without, ignore_index=True)

df_with['category'] = 'with'
df_without['category'] = 'without'

df = pd.concat([df_with, df_without], ignore_index=True)
df.head()
```

```python
plt.figure(figsize=(10, 6), dpi=300)

plt.gca().set_facecolor('white')
plt.gcf().set_facecolor('white')

for (category, file_path), group in df.groupby(['category', 'file_path']):
    color = 'red' if category == 'with' else 'blue'
    
    group['timestamp'] = group['timestamp'].apply(lambda t: dt.datetime.combine(dt.date.today(), t))
    plt.plot(group['timestamp'], group['hr'], label=category, color=color)

# Format the x-axis to display minutes and seconds
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
plt.gcf().autofmt_xdate()

# Optionally, remove minor ticks for cleaner look
# plt.gca().xaxis.set_minor_formatter(mdates.NullFormatter())

plt.xlabel('Time')
plt.ylabel('HR')
plt.title('HR with / without Polar H10')
plt.legend()
plt.show()
```
